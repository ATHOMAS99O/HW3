```python
import numpy as np
import matplotlib.pyplot as plt

#np.random.seed(1)  

p = 50
c_v = 20
c_s = 2
num_days = 1000

q_list = []
avg_profit_list = []
UB_list=[]
LB_list=[]


for q in range(1000,1100,1):                         #q = 1, 2, ..., 10

  #This is more efficient than the previous code because we are using an array instead of a for_loop 
  #to generaly daily demands for 'num_days'. 
  demand = np.random.uniform(501,1499,num_days)
  daily_profit = p * np.minimum(demand,q) + c_s * np.maximum(0,q-demand) - c_v * q
  
  avg_profit = np.mean(daily_profit)
  sample_var = np.var(daily_profit)
  half_width = 1.96 * np.sqrt(sample_var/num_days)
  q_list = q_list + [q]
  avg_profit_list = avg_profit_list +[avg_profit]
  UB_list=UB_list + [avg_profit + half_width]
  LB_list = LB_list + [avg_profit - half_width]

plt.plot(q_list, avg_profit_list,'o')
plt.plot(q_list,UB_list, '--.b',q_list,LB_list, '--.b', alpha=0.3)
plt.xlabel('q')
plt.ylabel('average profit')
plt.title('Newsvendor Problem (faster)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

num_darts = 100000

x = np.random.rand(num_darts)
y = np.random.rand(num_darts)
dart_position = ( (x-0.5)**2 + (y - 0.5)**2 ) <= 0.25

I_bar = np.mean(dart_position)
pi_hat = I_bar
half_width = 1.96*np.sqrt(I_bar*(1-I_bar)/(num_darts-1))
LB = pi_hat-half_width
UB = pi_hat+ half_width

plt.figure()
plt.plot(1, pi_hat, 'ok', 1, LB, '--.b', 1,UB, '--.c')
plt.plot(1, np.pi, '+r')
plt.gca().legend((r'estimate $\hat{\pi}$','LB of a 95% CI', 'UB of a 95% CI', r'True $\pi$'))
plt.ylim([2.5,3.8])
plt.xlabel('experiment count')
plt.ylabel(r'estimate of $\pi$')
plt.title(r'Estimation of $\pi$ over one expeirment')
plt.show()
```
